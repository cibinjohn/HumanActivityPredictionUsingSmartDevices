# Rounding the timestamp into seconds, Group wrt timestamp & extract mean value of the features

```python
smart_watch.timestamp = pd.to_datetime(smart_watch.timestamp)
smart_phone.timestamp = pd.to_datetime(smart_phone.timestamp)
smart_glasses.timestamp = pd.to_datetime(smart_glasses.timestamp)
```

```python
smart_watch['timestamp']=smart_watch['timestamp'].dt.round('1s')
smart_phone['timestamp']=smart_phone['timestamp'].dt.round('1s')
smart_glasses['timestamp']=smart_glasses['timestamp'].dt.round('1s')
```

```python
print("smart_watch shape : ",smart_watch.shape)
print("smart_phone shape : ",smart_phone.shape)
print("smart_glasses shape : ",smart_glasses.shape)
```

```python
smart_watch.groupby("timestamp").mean()
```

```python
smart_watch = smart_watch.groupby("timestamp").mean()
smart_glasses = smart_glasses.groupby("timestamp").mean()
smart_phone = smart_phone.groupby("timestamp").mean()
```

```python
smart_glasses.reset_index(inplace=True)
smart_watch.reset_index(inplace=True)
smart_phone.reset_index(inplace=True)
```

```python
print("smart_watch shape : ",smart_watch.shape)
print("smart_phone shape : ",smart_phone.shape)
print("smart_glasses shape : ",smart_glasses.shape)
```

```python
features_df = pd.merge(smart_watch, smart_glasses, on='timestamp', how='outer')
features_df = pd.merge(features_df, smart_phone, on='timestamp', how='outer')
```

```python
# Merging the features with labels
```

```python
report_df = pd.read_csv("final_report.csv")
report_df
```

```python
report_df['FROM'] = pd.to_datetime(report_df['FROM'])
report_df['TO'] = pd.to_datetime(report_df['TO'])
```

```python
# Create an empty DataFrame to store the expanded data
expanded_df = pd.DataFrame(columns=['timestamp', 'activity'])

# Iterate through each row in 'report_df' and expand the data for each second
for _, row in report_df.iterrows():
    from_time = row['FROM']
    to_time = row['TO']
    activity = row['activity']
    
    # Create a DatetimeIndex with 1-second frequency
    time_index = pd.date_range(from_time, to_time, freq='S')
    
    # Create a DataFrame with the expanded data for this row
    data = pd.DataFrame({'timestamp': time_index, 'activity': activity})
    
    # Append the data to the expanded DataFrame
    expanded_df = expanded_df.append(data)

# Reset the index of the expanded DataFrame
expanded_df.reset_index(drop=True, inplace=True)

# Print the resulting DataFrame
print(expanded_df)
```

```python
expanded_df.activity.value_counts()
```

```python
report_df = expanded_df
report_df.head()
```

```python
final_df = features_df.merge(report_df, on="timestamp", how="inner")
```
